## Метод linspace - генерує задану кількістю чисел у певному діапазоні, 50 чисел за замовчуванням

```python
import numpy as np

start = 1  # Початкове значення
size_of_vec = 100  # Кількість елементів у векторі

a = np.linspace(start, size_of_vec)
print(f"Згенерували {len(a)} чисел в діапазоні {a[0]} до {a[-1]}")
#print(f"Власне числа: {a}")

# Виведення перших та останніх 10 чисел
print("Перші 10 чисел:")
print(a[:10])

print("Останні 10 чисел:")
print(a[-10:])

# Визначення кроку
step = (a[-1] - a[0]) / (len(a) - 1)
print(f"Крок: {step}")

print("Кожне друге число:")
for i in range(0, len(a), 2):
    print(a[i])
homework("Вивів кожне друге число згенерованого діапазону :)")

print("Можна зменшити кількість чисел до 5:")
a = np.linspace(start, size_of_vec, 5)
print(a)

print("Розрахунок кроку без останнього значення:")
step_without_endpoint = (a[-1] - a[0]) / len(a)
print(f"Крок без останнього значення: {step_without_endpoint}")

homework("Збільшив кількість чисел до 100. Вивів перших та останніх 10.")

print("""Як бачимо перше та останнє значення завжди входять у згенерований діапазон. 
Відповідно при цьому вираховується крок.
Це можна змінити, виключивши останнє значення з визначення діапазону.
""")

print(np.linspace(start, size_of_vec, 5, endpoint=False))
homework("Порахував крок.")

_, sp_1 = np.linspace(start, size_of_vec, 5, endpoint=True, retstep=True)
_, sp_2 = np.linspace(start, size_of_vec, 5, endpoint=False, retstep=True)
print(f"Для порівняння, ми можемо визначити крок автоматично, у нашому випадку це {sp_1} та {sp_2}")
homework("Чому маємо такий крок?")
print(f"У першому прикладі, де ми встановили endpoint=True, крок визначається як різниця між останнім і першим значеннями, поділена на кількість елементів у векторі, включаючи останнє значення. ")
print(f"У другому прикладі, де ми встановили endpoint=False, крок визначається як різниця між останнім і першим значеннями, поділена на кількість елементів у векторі, виключаючи останнє значення.")
```

## Принцип Slicing може використовуватись для багатовимірних масивів.

```python
import numpy as np

A = np.array([
[11, 12, 13, 14, 15],
[21, 22, 23, 24, 25],
[31, 32, 33, 34, 35],
[41, 42, 43, 44, 45],
[51, 52, 53, 54, 55]])

print(f"""Масив:
{A}
Виміри: {A.shape}
Беремо декілька рядків: 
{A[0:2]}
Два перші стовпці: 
{A[:, 0:2]}
Два останні рядки:
{A[3:, :]}
Беремо підматрицю (нижні елементи): 
{A[-3:, -3:]}
""")
homework("Попрактикувався у витягуванні елементів, взяв матрицю 2х2 з нижнього та верхнього кутка матриці А.")

upper_left = A[:2, :2]  # верхній лівий кут
lower_right = A[-2:, -2:]  # нижній правий кут

print(f"Верхній лівий кут 2x2:\n{upper_left}")
print(f"Нижній правий кут 2x2:\n{lower_right}")
```

## Визначити за якою формулою генеруються значення для методу arange?
Метод arange генерує послідовність рівномірно розподілених чисел від початку до кінця заданого інтервалу. Формула для генерації таких чисел така:
'x = start + step * (i - 1)'

## Визначити формулу для методу linspace?
Метод linspace генерує послідовність рівномірно розподілених чисел від початку до кінця заданого інтервалу, включаючи кінцевий елемент. Формула для генерації таких чисел така:
'x = start + (end - start) * (i / (num - 1))'

## Порівняйте формули, для яких випадків вони можуть застосовуватись?
Основна відмінність між формулами для методів arange та linspace полягає в тому, що метод arange включає початкове значення інтервалу в послідовність, а метод linspace - ні.
Крім того, метод linspace дозволяє вказати кількість елементів послідовності, а метод arange - ні.
Застосування
Методи arange та linspace можна використовувати в різних випадках. Наприклад, вони можуть бути використані для створення послідовності чисел для графіка, для створення послідовності значень для ітераційного циклу або для створення послідовності значень для навчання моделі машинного навчання.

## Створіть декілька простих Векторів та Матриць та виведіть їх розмірність. Виконайте арифметичні операції над створеними Векторам/Матрицями;

```python
import numpy as np

# Створення векторів
vector_a = np.array([1, 2, 3])
vector_b = np.array([4, 5, 6])

# Виведення розмірності векторів
print("Розмір вектора A:", vector_a.shape)
print("Розмір вектора B:", vector_b.shape)

# Створення матриць
matrix_A = np.array([[1, 2, 3], [4, 5, 6]])
matrix_B = np.array([[7, 8, 9], [10, 11, 12]])

# Виведення розмірності матриць
print("Розмір матриці A:", matrix_A.shape)
print("Розмір матриці B:", matrix_B.shape)

# Арифметичні операції
sum_vector = vector_a + vector_b
difference_vector = vector_a - vector_b
elementwise_product_vector = vector_a * vector_b
dot_product_vector = np.dot(vector_a, vector_b)

sum_matrix = matrix_A + matrix_B
difference_matrix = matrix_A - matrix_B
elementwise_product_matrix = matrix_A * matrix_B
matrix_product = np.dot(matrix_A, matrix_B.T)  # Транспонуємо матрицю B перед множенням

# Виведення результатів
print("\nОперації з векторами:")
print("Сума:", sum_vector)
print("Різниця:", difference_vector)
print("Поелементний добуток:", elementwise_product_vector)
print("Добуток:", dot_product_vector)

print("\nОперації з матрицями:")
print("Сума:\n", sum_matrix)
print("Різниця:\n", difference_matrix)
print("Поелементний добуток:\n", elementwise_product_matrix)
print("Добуток матриць:\n", matrix_product)
```
